```python
def read_files():
    with open('./CUB_200_2011/bounding_boxes.txt') as f:
        lines = f.readlines()

    with open('./CUB_200_2011/images.txt') as f:
        images = f.readlines()
        
    return images, lines

def write_im(im):
    im_path = op.join(f"./birds/{phase}/images",f"{fname}.jpg")
    iio.imwrite(im_path,im)

def write_lbl(lbl):
    lbl_path = op.join(f"./birds/{phase}/labels",f"{fname}.txt")
    with open(lbl_path, "w") as text_file:
        text_file.write(lbl)

def process_im_txt(im_txt):
    imagedir = "./CUB_200_2011/images"
    im_path = op.join(imagedir, im_txt[:-1].split(" ")[1])
    im = iio.imread(im_path)
    return im
        
def process_bbox_txt(line):
    line = line.split(" ")
    line = line[1:]
    line.insert(0,"0")
    line = " ".join(line)
    line = line[:-1]
    return line
```

```python
images, lines = read_files()
phases = ["train","valid","test"]
total_idxs = np.random.choice(np.arange(11788), 1160)

for phase in phases:
    if phase == "train": idxs = total_idxs[:1000]
    elif phase == "valid": idxs = total_idxs[1000:1080]
    elif phase == "test": idxs = total_idxs[1080:1160]
        
    for idx in idxs:
        fname = f"calbird{idx}"

        im_txt = images[idx]
        bbox_txt = lines[idx]

        im = process_im_txt(im_txt)
        lbl = process_bbox_txt(bbox_txt)

        write_im(im)
        write_lbl(lbl)
```
